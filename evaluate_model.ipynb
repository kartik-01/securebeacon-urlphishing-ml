```python
from pathlib import Path
import json
import pandas as pd

metrics_path = Path('metrics.json')
with metrics_path.open() as fp:
    metrics = json.load(fp)
pd.Series(metrics, name='score')
```

```python
from fastapi.testclient import TestClient
from app import app

client = TestClient(app)
urls = [
    'https://www.paypal.com',
    'https://faecbook.com',
    'https://netflex-login.net',
    'https://google.com',
    'https://steamncommiunity.com'
]
rows = []
for url in urls:
    resp = client.post('/predict', json={'url': url})
    rows.append({**resp.json(), 'url': url})
pd.DataFrame(rows)
```
